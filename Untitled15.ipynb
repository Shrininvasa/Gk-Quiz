```python
import random 
questions = ["What is the capital of Canada?", "What is the capital of India?", "What is the capital of USA?",
             "What is the capital of China?", "What is the capital of Uganda?", "What is the capital of South Korea?",
             "What is the capital of North Korea?",
             "What is the capital of Japan?", "What is the capital of Germany?", "What is the capital of Australia?"]
answers = ["Ottawa", "New Delhi", "Washington DC",
           "Beijing", "Kampala", "Seoul",
           "Pyongyang", "Tokyo", "Berlin", "Canberra"]
marks = 0

def score_():
    print("Your Score is:", marks)
    
def remark_():
    remarks = ["Outstanding", "Outstanding", "Excellent", "Excellent", "Good", "Good", "Read more to score more",
               "Read more to score more",
              "Need to take interest", "General knowledge will always help you. Take it seriously"]
    print(remarks[10-marks])
    
for i in range(0,10):
    randominteger = random.randint(0,len(questions)-1)
    print(questions[randominteger])
    answer = input()
    if answer == answers[randominteger]:
        print("Correct!")
        marks += 1
    elif answer == "quit":
        break
    else:
        print("Wrong :(")
    
    questions.remove(questions[randominteger])
    answers.remove(answers[randominteger])
    
print("Please type what you would like to see (Score/Remark):")
i = input()
while i != "quit":
    if i == "Score":
        score_()
    if i == "Remark":
        remark_()
    print("Please type what you would like to see (Score/Remark):")
    i = input()
print(marks)
```
